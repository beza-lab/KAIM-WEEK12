```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the economic indicators dataset
economic_indicators_df = pd.read_csv('D:/KAIM/Week 12/KAIM WEEK12/Economic_Indicators.csv', parse_dates=['Date'])

# Load the Brent oil prices dataset
df_brent = pd.read_csv('D:/KAIM/Week 12/KAIM WEEK12/notebooks/Cleaned_BrentOilPrices.csv', parse_dates=['Date'])

# Merge the datasets on the Date
merged_df = pd.merge(df_brent, economic_indicators_df, left_on='Date', right_on='Date', how='inner')

# Display summary statistics
print(merged_df.describe())
```

```python
print("Columns in Economic Indicators Dataset:")
print(economic_indicators_df.columns)

print("\nColumns in Merged Dataset:")
print(merged_df.columns)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(merged_df['Date'], merged_df['Price'], label='Brent Oil Price', color='blue')
plt.plot(merged_df['Date'], merged_df['GDP_Growth'], label='GDP Growth', color='green')  
plt.title('Trends in Brent Oil Prices and GDP Growth Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(merged_df['Inflation_Rate'], merged_df['Price'], alpha=0.7, color='purple')
plt.title('Relationship Between Inflation Rate and Brent Oil Price')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Brent Oil Price')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(merged_df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Between Economic Indicators and Brent Oil Prices')
plt.show()
```
